```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/content/unemployment analysis.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df_melted = df.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Unemployment_Rate')
df_melted['Year'] = pd.to_numeric(df_melted['Year'])

avg_unemployment_per_country = df_melted.groupby('Country Name')['Unemployment_Rate'].mean().reset_index()
avg_unemployment_per_country = avg_unemployment_per_country.sort_values(by='Unemployment_Rate', ascending=False)

plt.figure(figsize=(15, 8))
sns.barplot(x='Country Name', y='Unemployment_Rate', data=avg_unemployment_per_country.head(10))
plt.title('Top 10 Countries by Average Unemployment Rate (1991-2021)')
plt.xlabel('Country Name')
plt.ylabel('Average Unemployment Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
avg_unemployment_over_time = df_melted.groupby('Year')['Unemployment_Rate'].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(avg_unemployment_over_time['Year'], avg_unemployment_over_time['Unemployment_Rate'])
plt.title("Average Unemployment Rate Over Time (1991-2021)")
plt.xlabel("Year")
plt.ylabel("Average Unemployment Rate (%)")
plt.grid(True)
plt.show()
```

```python
fig = px.line(df_melted, x='Year', y='Unemployment_Rate', title='Unemployment Trend Over Time')
fig.show()
```
